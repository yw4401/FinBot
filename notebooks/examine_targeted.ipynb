```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orca_rouge = pd.read_excel("gs://scraped-news-article-data-null/mistral-orca-rouge.xlsx")
sum_mis_rouge = pd.read_excel("gs://scraped-news-article-data-null/mistral-sum-rouge.xlsx")
```

```python
orca_rouge["model"] = "Orca"
sum_mis_rouge["model"] = "Tuned-Orca"

compiled_df = pd.concat([orca_rouge[["rouge2", "model"]], sum_mis_rouge[["rouge2", "model"]]],
                        ignore_index=True)
compiled_df.head()
```

```python
sns.barplot(compiled_df, x="model", y="rouge2", errorbar="ci", hue="model")
plt.title("Rouge2 Performance Summary")
plt.xlabel("Model")
plt.ylabel("Rouge-2")
plt.show()
```

```python
hallucination_df = pd.DataFrame({"F1": [0.31, 0.73], "Model": ["Orca", "Tuned-Orca"]})
sns.barplot(hallucination_df, x="Model", y="F1", errorbar="ci", hue="Model")
plt.title("Hallucination Evaluation Summary")
plt.show()
```
